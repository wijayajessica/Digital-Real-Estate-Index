## 2. Merge the data

### 2.1 joining house market listing property to assessor 

#### 2.1.1 using fips+apn


```python
# first using fips+apn
for i in assessor.columns:
    if 'apn' in i:
        print(i)    
```

```python
print(assessor.apnseqnbr.isnull().sum()/len(assessor))
print(assessor.oldapn.isnull().sum()/len(assessor))
print(assessor.formattedapn.isnull().sum()/len(assessor))
print(join.apn.isnull().sum()/len(join))
```

```python
#only the formattedapn column could be used
mergedapn = pd.merge(assessor, join, left_on='formattedapn',right_on='apn',how='inner')
mergedapn
```

#### 2.1.2 uding full street address + zipcode

```python
#Next using full_street_address + zipcode
for i in assessor.columns:
    if 'street' in i:
        print(i) 
for i in assessor.columns:
    if 'zip' in i:
        print(i) 
```

```python
for i in join.columns:
    if 'street' in i:
        print(i)  
for i in join.columns:
    if 'zip' in i:
        print(i) 
```

```python
merged_address = pd.merge(assessor, join, left_on=['situsstdfullstreetaddress','situsstdzip5'],right_on=['match_street_address','zipcode'],how='inner')
merged_address
```

### 2.2 joining assessor to transaction

```python
#drop duplicates
merged_address = merged_address.drop_duplicates('propertyid',keep='first')
trans = trans.drop_duplicates('propertyuniqueid',keep='first')
merged = pd.merge(merged_address, trans, left_on='propertyid',right_on='propertyuniqueid',how='inner')
merged 
```

```python
#Finally we get 427 columns!
#select informative columns and drop duplicates
#First drop uninformative columns with most of values nan
dropped_column = (merged.isnull().sum(axis=0)/9318)>0.9
merged = merged.drop(merged.columns[dropped_column],axis=1)
```

```python
print(merged.shape)
merged.head()
```

```python
merged = merged.T.drop_duplicates(keep='first').T
```

```python
del merged['Unnamed: 0_x']
del merged['Unnamed: 0_y']
```

```python
merged.to_csv('merged_first_american_'+fips+'.csv')
#merged
```
