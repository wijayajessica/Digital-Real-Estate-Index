```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
import sys
# /usr/local/anaconda3/bin/python !pip3 install pmdarima
# !{sys.executable} -m pip install pmdarima
import pmdarima as pm
```

## Load Data

```python
df_prop = pd.read_csv("../data/Boston-MLS/boston_property_info.csv")
df_listing = pd.read_csv("../data/Boston-MLS/boston_listing_dates.csv")
```

```python
display(df_prop.head())
display(df_listing.head())
```

```python
df = pd.merge(df_prop, df_listing.drop("zipcode", axis=1), left_on="cc_property_id", right_on="property_id", how="inner")
# df.to_csv("data/property_listings.csv")
```

```python
df.zipcode = df.zipcode.astype("str")
```

## EDA: Distribution per City, Zipcode

```python
dist_zip = df.dropna(subset=["city"]).groupby(["zipcode", "city"]).agg({"property_id":"count", "sale_price":"sum"}).reset_index()
dist_zip["city_zip"] = dist_zip["city"]+" 0"+dist_zip["zipcode"]
```

```python
fig, ax = plt.subplots(1,2, figsize=(15,7))
sns.barplot(y="city_zip", x= "property_id", data=dist_zip.sort_values(["property_id"], ascending=False).head(10), 
            orient="h", ax=ax[0])
# plt.title("Count of Property Sale")
# plt.show()

# plt.figure(figsize=(10,7))
sns.barplot(y="city_zip", x= "sale_price", data=dist_zip.sort_values(["sale_price"], ascending=False).head(10), 
            orient="h", ax=ax[1])
# plt.title("Sum of Market Value")
# plt.show()
plt.tight_layout()
```

```python
df_sale = df.dropna(subset=["city"]).groupby(["zipcode", "city", "sale_date"
                                             ]).agg({"property_id":"count", "sale_price":"sum"}).reset_index()
df_sale["city_zip"] = df_sale["city"]+" 0"+df_sale["zipcode"]
df_sale["sale_month"] = df_sale.sale_date.apply(lambda x :x[:-3])
df_sale = df_sale.rename(columns={"property_id":"count_sale", "sale_price":"total_value"})
```

```python
top10sales = list(df_sale.groupby("city_zip").sum().sort_values("total_value", ascending=False).head(10).reset_index().city_zip)
top10unitSold = list(df_sale.groupby("city_zip").sum().sort_values("count_sale", ascending=False).head(10).reset_index().city_zip)

df_sale_monthly = df_sale.groupby(["sale_month", "city_zip"]).agg({"count_sale":"sum", "total_value":"sum"}).reset_index()
```

```python
df_sale_monthly = df_sale_monthly.sort_values("total_value", ascending=False)
df_sale_monthly.head()
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(x="sale_month", hue="city_zip",y="total_value", data=df_sale_monthly[df_sale_monthly["city_zip"].isin(top10unitSold)], ci=None)
plt.xticks(rotation=90)
plt.title("Top 10 area with Highest Market Value")
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(x="sale_month", hue="city_zip",y="count_sale", data=df_sale_monthly[df_sale_monthly["city_zip"].isin(top10unitSold)], ci=None)
plt.xticks(rotation=90)
plt.title("Top 10 area with Highest Unit Sold")
plt.show()
```

## Baseline Forecast per city, zipcode

### 1. Forecast for Market Value (Total Dollar Price of all Transaction)Â¶

```python
n_periods = 6
fig, ax = plt.subplots(5,2,sharex=True, figsize=(16,30))
for idx, area in enumerate(top10sales):
    model=pm.auto_arima(df_sale_monthly[df_sale_monthly["city_zip"]==area].sort_values("sale_month")["total_value"][:-n_periods], 
                  seasonal=True, m=12)
    print(model)

    pred, CI = model.predict(6, return_conf_int=True)
    df_plot = df_sale_monthly[df_sale_monthly["city_zip"]==area].sort_values("sale_month")

    ax[idx//2, idx%2].plot(range(len(df_plot)), df_plot["total_value"], label="actuals")
    ax[idx//2, idx%2].plot(range(len(df_plot)-n_periods,len(df_plot)), pred, label="pred")
    ax[idx//2, idx%2].fill_between(range(len(df_plot)-n_periods,len(df_plot)),CI[:,0], CI[:,1], alpha=0.3, label=("95% CI"))
    ax[idx//2, idx%2].set_xticks(range(len(df_plot))[::5])
    ax[idx//2, idx%2].set_xticklabels(df_plot.sale_month[::5], rotation=90)
    ax[idx//2, idx%2].set_title("SARIMA forecast for "+area)
    ax[idx//2, idx%2].legend()
    
    
```

### 1. Forecast for Number of Units SoldÂ¶

```python
n_periods = 6
fig, ax = plt.subplots(5,2,sharex=True, figsize=(16,30))
for idx, area in enumerate(top10unitSold):
    model=pm.auto_arima(df_sale_monthly[df_sale_monthly["city_zip"]==area].sort_values("sale_month")["count_sale"][:-n_periods], 
                  seasonal=True, m=12)
    print(model)

    pred, CI = model.predict(6, return_conf_int=True)
    df_plot = df_sale_monthly[df_sale_monthly["city_zip"]==area].sort_values("sale_month")

    ax[idx//2, idx%2].plot(range(len(df_plot)), df_plot["count_sale"], label="actuals")
    ax[idx//2, idx%2].plot(range(len(df_plot)-n_periods,len(df_plot)), pred, label="pred")
    ax[idx//2, idx%2].fill_between(range(len(df_plot)-n_periods,len(df_plot)),CI[:,0], CI[:,1], alpha=0.3, label=("95% CI"))
    ax[idx//2, idx%2].set_xticks(range(len(df_plot))[::5])
    ax[idx//2, idx%2].set_xticklabels(df_plot.sale_month[::5], rotation=90)
    ax[idx//2, idx%2].set_title("SARIMA forecast for "+area)
    ax[idx//2, idx%2].legend()
    
```

## Baseline Forecast per City

```python
df_sale = df.dropna(subset=["city"]).groupby(["city", "sale_date"
                                             ]).agg({"property_id":"count", "sale_price":"sum"}).reset_index()
# df_sale["city_zip"] = df_sale["city"]+" 0"+df_sale["zipcode"]
df_sale["sale_month"] = df_sale.sale_date.apply(lambda x :x[:-3])
df_sale = df_sale.rename(columns={"property_id":"count_sale", "sale_price":"total_value"})

top10sales = list(df_sale.groupby("city").sum().sort_values("total_value", ascending=False).head(10).reset_index().city)
top10unitSold = list(df_sale.groupby("city").sum().sort_values("count_sale", ascending=False).head(10).reset_index().city)

df_sale_monthly = df_sale.groupby(["sale_month", "city"]).agg({"count_sale":"sum", "total_value":"sum"}).reset_index()
df_sale_monthly = df_sale_monthly.sort_values("total_value", ascending=False)
df_sale_monthly.head()
```

### 1. Forecast for Market Value (Total Dollar Price of all Transaction)Â¶

```python
n_periods = 6
fig, ax = plt.subplots(5,2,sharex=True, figsize=(16,30))
for idx, area in enumerate(top10sales):
    model=pm.auto_arima(df_sale_monthly[df_sale_monthly["city"]==area].sort_values("sale_month")["total_value"][:-n_periods], 
                  seasonal=True, m=12)
    print(model)

    pred, CI = model.predict(6, return_conf_int=True)
    df_plot = df_sale_monthly[df_sale_monthly["city"]==area].sort_values("sale_month")

    ax[idx//2, idx%2].plot(range(len(df_plot)), df_plot["total_value"], label="actuals")
    ax[idx//2, idx%2].plot(range(len(df_plot)-n_periods,len(df_plot)), pred, label="pred")
    ax[idx//2, idx%2].fill_between(range(len(df_plot)-n_periods,len(df_plot)),CI[:,0], CI[:,1], alpha=0.3, label=("95% CI"))
    ax[idx//2, idx%2].set_xticks(range(len(df_plot))[::5])
    ax[idx//2, idx%2].set_xticklabels(df_plot.sale_month.unique()[::5], rotation=90)
    ax[idx//2, idx%2].set_title("SARIMA forecast for "+area)
    ax[idx//2, idx%2].legend()
```

### 2. Forecast for Number of Units Sold

```python
n_periods = 6
fig, ax = plt.subplots(5,2,sharex=True, figsize=(16,30))
for idx, area in enumerate(top10unitSold):
    model=pm.auto_arima(df_sale_monthly[df_sale_monthly["city"]==area].sort_values("sale_month")["count_sale"][:-n_periods], 
                  seasonal=True, m=12)
    print(model)

    pred, CI = model.predict(6, return_conf_int=True)
    df_plot = df_sale_monthly[df_sale_monthly["city"]==area].sort_values("sale_month")

    ax[idx//2, idx%2].plot(range(len(df_plot)), df_plot["count_sale"], label="actuals")
    ax[idx//2, idx%2].plot(range(len(df_plot)-n_periods,len(df_plot)), pred, label="pred")
    ax[idx//2, idx%2].fill_between(range(len(df_plot)-n_periods,len(df_plot)),CI[:,0], CI[:,1], alpha=0.3, label=("95% CI"))
    ax[idx//2, idx%2].set_xticks(range(len(df_plot))[::5])
    ax[idx//2, idx%2].set_xticklabels(df_plot.sale_month[::5], rotation=90)
    ax[idx//2, idx%2].set_title("SARIMA forecast for "+area)
    ax[idx//2, idx%2].legend()
    
```
